```python
import numpy as np

print("Hello!")

array = np.random.randint(1, 20, size=(5,4))
print(array)

rows, cols = array.shape
half_cols = cols // 2

array1 = array[:, :half_cols]
array2 = array[:, half_cols:]

print("\n первая часть массива: ")
print(array1)

print("\n первая часть массива: ")
print(array2)

x = int(input("\n Введите число: "))
count = 0

for i in range(len(array1)):
  for j in range(len(array1[i])):
    if array1[i, j] == x:
      count += 1

if count > 0:
  print(f"Число {x} встретилось {count} раз")
if count == 0:
  print(f"Число {x} не встретилось")


```

```python
import pandas as pd
film = pd.read_csv("/content/flixpatrol.csv", delimiter=",", nrows = 1000)
film.head()
```

```python
#средний
film["Rank"].mean()

film['Watchtime'] = film['Watchtime'].astype(str)
film['Watchtime'] = film['Watchtime'].str.replace(',', '')
film['Watchtime'] = film['Watchtime'].astype(float)
film['Watchtime'] = film['Watchtime'].round(2)

genre_watchtime = film.groupby('Genre')['Watchtime'].sum().sort_values(ascending=False)
print("\nАнализ времени просмотра по жанрам:")
print(genre_watchtime)
############################

top_genre = film.groupby('Genre').size().sort_values(ascending=False).head(5)
print("\nТоп 5 популярных жанров:")
print(top_genre)
################################

watchtime_grouped = film.groupby('Watchtime').size().sort_values(ascending=True)
print('\nГруппировка времени просмотра по убыванию:')
print(watchtime_grouped)
#################################

drama_movies = film[film['Genre'] == 'Drama']
drama_movies_grouped = drama_movies.groupby('Title')['Watchtime in Million'].sum().sort_values(ascending=False).head(5)
print('\nТоп-5 самых просматриваемых фильмов жанра "драма":')
print(drama_movies_grouped)

```
