```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
cards = pd.read_csv('/content/Credit_card.csv');
cards_label = pd.read_csv('/content/Credit_card_label.csv');
```

```python
cards_label.head()
```

```python
cards_label.dtypes
cards.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cards = cards.apply(le.fit_transform)
cards.head()
```

```python
y = cards_label['label']
X = cards
y.head()
```

```python
from sklearn.model_selection import cross_val_score, train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)

print("Inputs X (train/test):", X_train.shape, X_test.shape)

print("Outputs Y (train/test):", y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Инициализация random_state обеспечит воспроизводимость результатов
tree = DecisionTreeClassifier(random_state=0)

# построения модели
tree.fit(X_train, y_train)

# Оценить качество модели можно, вычислив правильность (accuracy):

print("Правильность на обучающем наборе: {}".format(tree.score(X_train, y_train)))
print("Правильность на тестовом наборе: {}".format(tree.score(X_test, y_test)))
```

```python
tree = DecisionTreeClassifier(max_depth=10, random_state=0)
tree.fit(X_train, y_train)

print("Правильность на обучающем наборе: {:.3f}".format(tree.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
print("-------------Prediction------------")

tree_valid_pred = tree.predict(X_test) # прогноз на отложенной выборке
print(X_test.head(10))
print(tree_valid_pred[0:10]) # первые 10 спрогнозированых меток
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

ac_score = accuracy_score(y_test, tree_valid_pred)
print("Accuracy score:", ac_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X_train = np.ascontiguousarray(X_train)
y_train = np.ascontiguousarray(y_train)

first_knn = KNeighborsClassifier()
knn_model = first_knn.fit(X_train, y_train)
knn_model.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
X_test = np.ascontiguousarray(X_test)
knn_valid_pred = knn_model.predict(X_test)
confusion_mat = confusion_matrix(y_test, knn_valid_pred)
print("Confusion matrix:")
print(confusion_mat)
```
